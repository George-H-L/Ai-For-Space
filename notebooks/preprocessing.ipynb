# Preprocessing - SAT-6 Dataset

**Owner:** Ehsan

## Purpose
Load the SAT-6 dataset, perform preprocessing, and produce train/validation/test splits saved to the shared Google Drive folder for use by the rest of the team.

## Steps
1. Mount Google Drive
2. Load raw SAT-6 data
3. Class mapping and label encoding
4. Normalisation
5. Data augmentation (training set only)
6. Train/validation/test split (fixed seed)
7. Save processed splits to shared Drive folder
8. Build PyTorch dataloaders
