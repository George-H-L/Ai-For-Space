# Evaluation & Comparison

**Owners:** George & Tausif

## Purpose
Evaluate trained models on the test set, compute comparison metrics, and produce visualisations for the report.

## Steps
1. Load preprocessed test set from shared Drive
2. Load trained model weights (ResNet18, ResNet50)
3. Compute metrics: accuracy, precision, recall, F1-score (TBC)
4. Generate confusion matrices
5. Per-class performance breakdown
6. Visualisations and comparison plots
7. Export results for the report
